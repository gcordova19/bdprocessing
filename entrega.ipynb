```python
# File location and type
file_location = "/FileStore/tables/world_happiness_report_2021.csv"
file_type = "csv"

# CSV options
first_row_is_header = "true"
delimiter = ","
#define el data schema del archivo que queremos leer
purchaseSchema = StructType([
    StructField("Country name", StringType(), True),
    StructField("Regional indicator", StringType(), True),
    StructField("Ladder score", FloatType(), True),
    StructField("Standard error of ladder score", FloatType(), True),    
    StructField("upperwhisker", FloatType(), True),
    StructField("lowerwhisker", FloatType(), True),
    StructField("Logged GDP per capita", FloatType(), True),
    StructField("Social support", FloatType(), True),  
    StructField("Healthy life expectancy", FloatType(), True),
    StructField("Freedom to make life choices", FloatType(), True),
    StructField("Generosity", FloatType(), True),
    StructField("Perceptions of corruption", FloatType(), True),
    StructField("Ladder score in Dystopia", FloatType(), True),
    StructField("Explained by: Log GDP per capita", FloatType(), True),
    StructField("Explained by: Social support", FloatType(), True),
    StructField("Explained by: Healthy life expectancy", FloatType(), True),
    StructField("Explained by: Freedom to make life choices", FloatType(), True),
    StructField("Explained by: Generosity", FloatType(), True),
    StructField("Explained by: Perceptions of corruption", FloatType(), True),
    StructField("Dystopia + residual", FloatType(), True),
])  
# The applied options are for CSV files. For other file types, these will be ignored.
df_report_2021 = spark.read.format(file_type) \
  .option("schema", purchaseSchema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df_report_2021)
```

- "Ladder score in Dystopia": una referencia a la puntuación de felicidad en una situación imaginaria de "distopía".
- "Explained by: Log GDP per capita": la contribución del PIB per cápita a la puntuación de felicidad.

2. ¿Cuál es el país más “feliz” del 2021 por continente según la data?

```python
display(df_report_2021)
```

```python

# Agrupar los datos por continente y obtener el valor máximo de "Ladder score"
df_maxi_continent = df_report_2021.groupBy("Regional indicator") \
    .agg(max("Ladder score").alias("Max Ladder score"))

df_maxi_continent.printSchema()
df_maxi_continent = df_maxi_continent.withColumnRenamed("max(Ladder score)", "max_ladder_score") \
    .withColumnRenamed("Regional indicator", "regional_indicator")

display(df_maxi_continent)

# Unir el DataFrame anterior con el DataFrame original para obtener el nombre del país correspondiente al valor máximo de "Ladder score" por continente
df_pais_feliz_continent = df_report_2021.join(df_maxi_continent, 
                                                       (df_report_2021["Regional indicator"] == df_maxi_continent["regional_indicator"]) & 
                                                       (df_report_2021["Ladder score"] == df_maxi_continent["Max Ladder score"]))

df_pais_feliz_continent.printSchema()
display(df_pais_feliz_continent)
# Seleccionar las columnas de interés y mostrar los resultados
df_pais_feliz_continent.select("Regional indicator", "Country name", "Ladder score").show()
```

```python
df_report_2021 = df_report_2021.withColumnRenamed("Ladder score", "Life Ladder") \
    .withColumnRenamed("Logged GDP per capita", "Log GDP per capita") \
    .withColumnRenamed("Healthy life expectancy", "Healthy life expectancy at birth")
df_report_2021 = df_report_2021.withColumn('year', lit('2021'))
df_report_2021 = df_report_2021.withColumn("Positive affect", lit(None).cast('Float'))
df_report_2021 = df_report_2021.withColumn("Negative affect", lit(None).cast('Float'))
df_report_2021 = df_report_2021.select('Country name', 'year', 'Life Ladder',  'Log GDP per capita', 'Social support', 'Healthy life expectancy at birth', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption','Positive affect', 'Negative affect', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual', 'Regional indicator')
df_report_2021.printSchema()
display(df_report_2021)

df_report = df_report.withColumn('Standard error of ladder score', lit(None).cast('string')) 
df_report = df_report.withColumn('upperwhisker', lit(None).cast('Float')) 
df_report = df_report.withColumn('lowerwhisker', lit(None).cast('Float')) 
df_report = df_report.withColumn('Ladder score in Dystopia', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Log GDP per capita', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Social support', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Healthy life expectancy', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Freedom to make life choices', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Generosity', lit(None).cast('Float')) 
df_report = df_report.withColumn('Explained by: Perceptions of corruption', lit(None).cast('Float'))  
df_report = df_report.withColumn('Dystopia + residual', lit(None).cast('Float')) 
df_report = df_report.withColumn('Regional indicator', lit(None).cast('Float'))  
df_report.printSchema()
display(df_report)

merged_df = df_report_2021.union(df_report)
display(merged_df)
merged_df.select("year").distinct().show()

# Obtener los valores únicos de la columna 'year'
years_list = merged_df.select("year").distinct().rdd.flatMap(lambda x: x).collect()

# Imprimir la lista de años únicos
print(years_list)
top_country_dict = {}

for year in years_list:
    # Sub-dataframe agrupado por año y ordenado por Life Ladder
    sub_df = merged_df.filter(merged_df.year == year)#.groupBy("year").agg({"Life Ladder": "max"})
    sub_df = sub_df.orderBy("Life Ladder", ascending=False)
    
    # Guarda la primera fila en un diccionario
    top_country_dict[year] = {
        "Country": sub_df.first()["Country name"],
        "Life Ladder": sub_df.first()["Life Ladder"]
    }
    
    display(sub_df)
print(top_country_dict)

country_count = {}

for year in top_country_dict:
    country = top_country_dict[year]['Country']
    if country in country_count:
        country_count[country] += 1
    else:
        country_count[country] = 1

print(country_count)

from functools import reduce

# Assuming you have a dictionary called country_count
max_count = reduce(lambda x, y: x if x > y else y, country_count.values())

# Filter the dictionary for keys that have the max_count value
max_countries = [k for k, v in country_count.items() if v == max_count]

# Print the country/countries with the maximum count
print("Los paises o pais durante estos años que se ha repetido mas son/ es:", max_countries)


```
